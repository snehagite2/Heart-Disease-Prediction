**This code loads a dataset, handles missing values, and saves the cleaned dataset**

**pandas:** Used for data manipulation and handling tabular data.

**numpy:** Used for numerical operations (though not explicitly used in this code).

If missing values exist, it fills them with the median of the respective column using **df.fillna(df.median(), inplace=True)**.

If no missing values are found, it prints **"No missing values found."**

The dataset already had no missing values.                                     
No changes were made, but the dataset was still saved

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = "/content/heart.csv"
df = pd.read_csv(file_path)

# Handle missing values
def handle_missing_values(df):
    print("Missing values before handling:")
    print(df.isnull().sum())

    if df.isnull().sum().sum() > 0:
        df.fillna(df.median(), inplace=True)
        print("Missing values handled using median imputation.")
    else:
        print("No missing values found.")

handle_missing_values(df)

# Save the cleaned dataset
df.to_csv("/content/heart.csv", index=False)
print("Missing values preprocessing completed. Saved as heart_missing_values_handled.csv")
```

This code generates histograms for all numerical columns in a DataFrame (df) to visualize the distribution of each attribute.

**df.hist()** is a Pandas built-in method to create histograms for numerical columns.                                                                                   
Histograms show the distribution of data by grouping values into bins and counting the number of occurrences.

**bins=30** → Divides the data range into 30 intervals (bins) for each histogram.        
**layout=(4,4) **→ Creates a 4x4 grid (4 rows, 4 columns) for displaying multiple histograms.                                                                         
**figsize=(40,30)**→ Sets the figure size (40 inches wide, 30 inches tall) for better visualization.                                                            
**column=df.columns[0:]** → Selects all columns for plotting histograms.             
**color="DarkCyan"**→ Sets the histogram color to dark cyan.                        
**grid=False** → Removes grid lines for a cleaner look.        

```python
# Histogram to check the distribution of each attribute
import pandas as pd
import numpy as np
import seaborn as sns # Importing necessary libraries
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/heart.csv"
df = pd.read_csv(file_path) # Re-defining df in this cell

# Histogram to check the distribution of each attribute

df.hist(bins = 30 , layout = (4,4), figsize = (10,20)  ,column = df.columns[0:], color= "DarkCyan", grid = False)
plt.show()
```

This code calculates and visualizes the correlation between numerical variables in a dataset using a heatmap.

This will generate a heatmap where:

Cells are color-coded based on correlation strength.
Values inside each cell indicate the correlation coefficient.

Darker blue-purple shades → Weak or negative correlation.                           
Brighter shades → Strong positive correlation.                                    
Example interpretations:                                                            
"Age" and "Thalach" have a negative correlation (-0.39) → Older people tend to have lower maximum heart rate.                                                    
"Thalach" and "Target" have a positive correlation (0.42) → Higher heart rate is linked with heart disease.                                                        
"Chol" and "Target" have a weak correlation (-0.09) → Cholesterol alone may not strongly indicate heart disease.

```python
df_corr= df.corr()
plt.figure(figsize = (14,5))
sns.heatmap(df_corr,annot=True,cmap = 'mako')
```

This boxplot visualization helps in data exploration, outlier detection, and feature analysis.

Defines a list of numerical columns for which boxplots will be created.          
These features are continuous variables in a heart disease dataset:                
**age** → Patient's age.                                                                  
**trestbps** → Resting blood pressure.                                               
**chol** → Serum cholesterol level.                                                    
**thalach** → Maximum heart rate achieved.                                            
**oldpeak** → ST depression induced by exercise                                                                                   
                                                                 Sets the Seaborn style to "whitegrid", which provides a clean background with grid lines.

Creates a separate figure (plt.figure(figsize=(6, 4))) for each column to avoid overlapping plots.

sns.boxplot(x=df[column], palette="Set2")

x=df[column] → Plots the column's values on the x-axis.

palette="Set2" → Uses a soft color palette for better visuals

```python
numerical_columns= ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
# List of (by default) non-categorical attributes

#Boxplot

sns.set(style="whitegrid")

# Iterate through the numerical columns and create a boxplot for each one
for column in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column], palette="Set2")
    plt.title(f'Boxplot of {column}', fontsize=16, fontweight='bold', color='darkblue')
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Value', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
